```python
class ValueIteration(object):
    def __init__(self, transitionTable, rewardTable, valueTable, convergenceTolerance, gamma):
        self.transitionTable = transitionTable
        self.rewardTable  = rewardTable
        self.valueTable = valueTable
        self.convergenceTolerance = convergenceTolerance
        self.gamma = gamma

    def __call__(self):
        #######################################
        ########## YOUR CODE HERE #############
        #######################################
        transition_key=[]
        for i in (self.transitionTable.keys()):
            transition_key.append(i)
    
        rewardTableProb_key=[]
        for i in (self.rewardTable.keys()):
            rewardTableProb_key.append(i)
    
        valueTableProb_key= []
        for i in (self.valueTable.keys()):
            valueTableProb_key.append(i)
    
        def getKeysByValue(dictOfElements, valueToFind):
            listOfKeys = list()
            listOfItems = dictOfElements.items()
            for item  in listOfItems:
                if item[1] == valueToFind:
                    listOfKeys.append(item[0])
            return listOfKeys

        def one_step_lookahead(state, V,R,T):
            A = {}
            for action in T[state]:
                A[action] =0
                for next_state , prob in T[state][action].items():
                    A[action] += prob * (R[state][action][next_state] + self.gamma * V[next_state])
            return A
        ValueTable=self.valueTable
        while True:
            # Stopping condition
            delta = 0
            # Update each state...
            for s in valueTableProb_key:
                # Do a one-step lookahead to find the best action
                A = one_step_lookahead(s, ValueTable,self.rewardTable,self.transitionTable)
                B=[]
                for i in (A.values()):
                    B.append(i)
                best_action_value = np.max(B)
                # Calculate delta across all states seen so far
                delta = max(delta, np.abs(best_action_value - ValueTable[s]))
                # Update the value function. Ref: Sutton book eq. 4.10. 
                ValueTable[s] = best_action_value        
            # Check if we can stop 
            if delta < self.convergenceTolerance:
                break
        # Create a deterministic policy using the optimal value function
        policyTable = {} 
        for s in valueTableProb_key:
        # One step lookahead to find the best action for this state
            A = one_step_lookahead(s, ValueTable,self.rewardTable,self.transitionTable)
            B =[]
            for i in (A.values()):
                B.append(i)
            best_action_value = np.max(B)
            best_action = getKeysByValue(A,best_action_value)
            p=[]
            for v in self.transitionTable[s][best_action[0]].values():
                p.append(v)
        # Always take the best action
            policyTable[s] = {best_action[0]:p[0]}
        gamma = .95
        beta = .4
        optimalQA = {}
    
        for s in valueTableProb_key:
        # One step lookahead to find the best action for this state
            A = one_step_lookahead(s, ValueTable,self.rewardTable,self.transitionTable)
            Q=[]
            B =[]
            for i in (A.values()):
                B.append(i)        
            for i in range(len(B)):
                B[i] = np.exp(beta*B[i])
            for i in B:
                Q.append (i)
            for i in range(len(Q)):
                Q[i]=(Q[i])/(np.sum(B))
            best_action_pro = np.max(Q)
            optimalQA[s] = best_action_pro    
        return ValueTable, policyTable,optimalQA
```
